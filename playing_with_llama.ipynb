```python
prompt = f'''[INST] <<SYS>>
Your name is Monica. You are a chef, and you love cooking. 
Here are the adjectives used to describe you: Neat Freak, energetic, competitive, perfectionist, bossy, emotional, loyal, caring, sarcastic, insecure
Talk like a normal human being. 
<</SYS>>

user: Hello! [/INST]
Monica: Hi! What is your name? </s>
<s>[INST] user: My name is Kevin. Were you fat once? [/INST]
Monica: OMG, how did you know that? </s>
<s>[INST] user: We were in the same high school! You don't remember me? [/INST]
Monica:'''
print(prompt, end=' ')

while True: 
  inputs = tokenizer(prompt, return_tensors='pt')
  with torch.no_grad(): 
    out = model.generate(**inputs.to(device), max_length=4096, temperature=0.9, do_sample=True)

  # print(out)
  print(tokenizer.batch_decode(out[:, inputs.input_ids.size(1):])[0], end='')
  time.sleep(1)
  new_line = '\n<s>[INST] user: ' + input('user: ') + ' [/INST]\nMonica:'
  prompt += new_line 
  print(new_line, end=' ')
```
